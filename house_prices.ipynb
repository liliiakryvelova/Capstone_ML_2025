```python
import pandas as pd
```

```python
df = pd.read_csv("~/Course_Python/data_splitting/Capstone_ML_2025/house_prices.csv")
df
```

```python
# Get the number of rows and columns
num_rows, num_columns = df.shape

# Get the number of unique columns
num_unique_columns = len(df.columns)

print(f"Number of rows: {num_rows}")
print(f"Number of unique columns: {num_unique_columns}")
```

```python
# Check for NULL values in each column
null_values = df.isnull().sum()

# Filter out columns with no NULL values
null_columns = null_values[null_values > 0]

print("Columns with NULL values and the count of NULLs:")
print(null_columns)
```

```python
# Generate summary statistics using describe()
summary_stats = df.describe()

# Display the summary statistics
print(summary_stats)
```

```python
# Drop rows with any NULL values
df_cleaned = df.dropna()

# Display the number of rows and columns after dropping NULL values
print(f"Shape of the cleaned dataset: {df_cleaned.shape}")

# Optionally, display the first few rows of the cleaned dataset
print(df_cleaned.head())
```
